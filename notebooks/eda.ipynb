```python
import pandas as pd

# Load data (replace with your dataset)
import os
print(os.path.exists(r"C:\10x AIMastery\Predicting-Price-Moves-with-News-Sentiment\Data\raw_analyst_ratings.csv\raw_analyst_ratings.csv"))

df = pd.read_csv(r"C:\10x AIMastery\Predicting-Price-Moves-with-News-Sentiment\Data\raw_analyst_ratings.csv\raw_analyst_ratings.csv")

# Text length stats
df["headline_length"] = df["headline"].apply(len)
print(df["headline_length"].describe())

# Articles per publisher
print(df["publisher"].value_counts().head(10))
```

```python
# Convert dates to datetime
df["date"] = pd.to_datetime(df["date"], format='ISO8601')

# Plot articles per day
df.set_index("date").resample("D").size().plot(title="Daily Articles")
```

```python
from nltk import FreqDist

# Extract keywords
words = " ".join(df["headline"]).split()
freq = FreqDist(words)
print(freq.most_common(10))
```
